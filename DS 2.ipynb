```python
Q1 = df['Sem1'].quantile(0.25)
Q3 = df['Sem1'].quantile(0.75)
print("Q1: ",Q1)
print("Q3: ",Q3)
```

```python
IQR=Q3-Q1
IQR
```

```python
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
print("Upper limit: ",UL)
print("lower Limit: ",LL)
```

# Finding Outliers

```python
df[df['Sem1']>UL]
```

```python
newdf=df[df['Sem1']<UL]
newdf
```

```python
plt.figure(figsize=(16,5))

plt.subplot(2,2,1)
plt.boxplot(df['Sem1'])
plt.title('Befor')

plt.subplot(2,2,2)
plt.boxplot(newdf['Sem1'])
plt.title('After')

plt.show()
```

```python
q1 = df['average Score'].quantile(0.25)
q3 = df['average Score'].quantile(0.75)
print("Q1: ",q1)
print("Q3: ",q3)
```

```python
df['average Score'].describe()
```

```python
iqr=q3-q1
iqr
```

# Finding Outliers

# Handling outliers

```python
new_df=df[df['average Score']<Upper]
new_df
```

```python
plt.figure(figsize=(16,5))

plt.subplot(2,2,1)
plt.boxplot(df['average Score'])
plt.title('Befor')

plt.subplot(2,2,2)
plt.boxplot(new_df['average Score'])
plt.title('After')

plt.show()
```

```python
H=df['Sem2'].mean() + 3*df['Sem2'].std()
L=df['Sem2'].mean() - 3*df['Sem2'].std()
print("highest: ",H)
print("Lowest: ",L)
```

```python
df[(df['Sem2']>H) | (df['Sem2']<L) ]
```
