```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("iris")
```

```python
df
```

```python
fig, axs=plt.subplots(2,2,figsize=(16,10))

sns.boxplot(x='species',y='sepal_length', data=df, ax=axs[0,0])
sns.boxplot(x='species',y='sepal_width', data=df, ax=axs[0,1])
sns.boxplot(x='species',y='petal_length', data=df, ax=axs[1,0])
sns.boxplot(x='species',y='petal_width', data=df, ax=axs[1,1])
```

```python
fig, axs=plt.subplots(2,2,figsize=(16,10))

sns.histplot(df['sepal_length'], ax=axs[0,0])
sns.histplot(df['sepal_width'], ax=axs[1,0])
sns.histplot(df['petal_length'], ax=axs[0,1])
sns.histplot(df['petal_width'], ax=axs[1,1])
```

```python
fig, axs=plt.subplots(2,2,figsize=(16,10))

sns.distplot(df['sepal_length'], ax=axs[0,0])
sns.distplot(df['sepal_width'], ax=axs[1,0])
sns.distplot(df['petal_length'], ax=axs[0,1])
sns.distplot(df['petal_width'], ax=axs[1,1])
```
